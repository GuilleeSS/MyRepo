```python
def babylonian_sqrt(S, tolerance=1e-7, max_iterations=1000):
    # Initial guess
    x = S if S != 0 else 1.0

    for i in range(max_iterations):
        next_x = 0.5 * (x + S / x)

        # Check convergence
        if abs(next_x - x) < tolerance:
            break

        x = next_x

    return x


def check_estimation(S, approx):
    real_value = S ** 0.5

    if approx > real_value:
        return "overestimate"
    elif approx < real_value:
        return "underestimate"
    else:
        return "exact"


def main():
    S = float(input("Enter a non-negative number S: "))
    tolerance = float(input("Enter desired precision (e.g., 1e-7): "))

    result = babylonian_sqrt(S, tolerance)
    estimation_type = check_estimation(S, result)

    print(f"\nApproximated sqrt({S}) = {result}")
    print(f"Actual sqrt({S}) = {S**0.5}")
    print(f"The result is an {estimation_type}.")


if __name__ == "__main__":
    main()
```
